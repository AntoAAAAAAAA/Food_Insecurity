```python
###Important
'''Turns out that due to methodological changes in 2020, data prior to 2020 cannot 
be compared to data gathered from 2020 and beyond'''
'''This means we can either two do different analyses, or just stick with 2020 and beyond (which
would technically just be 1 dataset)'''
```

```python
t18_data = t18_data.rename(columns=t18_data.iloc[0]).drop(t18_data.index[0]).reset_index(drop=True)

"""
Cell generated by Data Wrangler.
"""
def clean_data(t18_data):
    # Drop column: '2018 Cost Per Meal'
    t18_data = t18_data.drop(columns=['2018 Cost Per Meal'])
    # Drop column: '2018 Weighted Annual Food Budget Shortfall'
    t18_data = t18_data.drop(columns=['2018 Weighted Annual Food Budget Shortfall'])
    return t18_data

t18_data_clean = clean_data(t18_data.copy())
t18_data_clean
```

```python
t18_data_clean['County, State'] = t18_data_clean['County, State'].str.replace(r',.*', '',regex=True)
t18_data_clean['County, State'] = t18_data_clean['County, State'].str.replace(r'County.*', '',regex=True)
t18_data_clean = t18_data_clean[t18_data_clean['State'] == 'TX']
t18_data_clean

### 2018 Data is clean
```

```python
t19_23_data['County, State'] = t19_23_data['County, State'].str.replace(r' County.*','',regex=True)
t19_23_data = t19_23_data[t19_23_data['State'] == 'TX']
t19_23_data = t19_23_data.drop(columns=['FIPS', 'Rural-Urban Continuum Code (2013)', 'Rural-Urban Continuum Code (2023)', 'Census Region', 'Census Division', 'FNS Region']).copy()
t19_23_data

### 2019 to 2023 Data is cleaned out 
```
